1. Drop the null values from the column "Book Name"

```python
df.dropna(subset=["Book Name"] ,inplace = True)
df.isna().sum()
```

2.Distribution of books by Publishing Year 


```python

plt.hist(df["Publishing Year"] , bins=20, edgecolor='black')
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Books by Publishing Year")
plt.show()
```

3. Number of books in each genre 


```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of books in each genre")
plt.show()
```

4. Top 5 authors on the basis of Book_Average_Rating

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False).head()
```

5.Book Ratings Count by Genre

```python
sns.boxplot(x = "genre", y = "Book_ratings_count",data = df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Boxplot of Book Ratings Count by Genre")
plt.xticks(rotation = 45)
plt.show()
```

9.Creating a scatter plot to visualize the relationship between Book Average Rating  and Book Rating Count

10.The total gross sales by authors

```python

total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
total_gross_sales_by_author
```

11.Top 5 authors according to gross sales

```python
total_gross_sales_by_author.sort_values(ascending = False).head()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head().plot(kind = "bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales by Each Author")
plt.show()
```

12.Units Sold for Each Author Rating

```python

sns.boxplot(x = "Author_Rating" , y = "units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Boxplot of Units Sold For Each Author Rating")
plt.show()
```

13.Total Books Sold over the years 

```python
df.groupby("Publishing Year")["units sold"].mean().plot(kind ="line", marker = "o")
plt.xlabel("Publishing Year")
plt.ylabel("Units Sold")
plt.title("Total Books Sold Over The Years")
plt.show()
```

14.Aggregate the data by genre

```python

df.groupby("genre").agg(
    Total_Sales=("gross sales","sum"),
    Avg_Rating=("Book_average_rating","mean"),
    Max_Price=("sale price","max"),
    Book_Count=("Book Name","count")
)
```
