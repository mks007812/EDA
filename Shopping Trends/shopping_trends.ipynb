4.Subscription Status by Pie chart

5.Color Distribution by Pie Chart

6.Top 5 colors

```python
good["Color"].value_counts().head().plot(kind = "bar")
plt.title("Bar Chart of Colors")
plt.xticks(rotation = 45)
plt.show()
```

7.Previous purchases vs review rating scatter plot

```python

good.plot(x = "Previous Purchases", y ="Review Rating" , kind = "scatter")
plt.title("Scatter Plot of Previous Purchases vs Review Rating")
plt.show()
```

```python
good.groupby("Previous Purchases")["Review Rating"].mean().plot(kind = "bar")
plt.xticks(rotation = 0, ha = "center")
plt.title("Bar Chart of Average Review Rating by Previous Purchases")
plt.show()
```

8.Seasonal Purchase Amount

```python
good.groupby("Season")["Purchase Amount (USD)"].sum().sort_values(ascending = False).plot(kind = "barh")
```

```python
good.groupby("Season")["Purchase Amount (USD)"].sum().sort_values(ascending = False).plot(kind = "bar")
plt.xticks(rotation = 0 )
plt.xlabel("Season")
plt.ylabel("Total Purchase Amount (USD)")
plt.title("Bar Chart of Total Purchase Amount of Seasons")
plt.show()
```

9.Purchase Amount (USD) by Frequency of Purchases

```python
good.boxplot(column = "Purchase Amount (USD)", by = "Frequency of Purchases")
plt.title("Boxplot of Purchase Amount (USD) by Frequency of Purchases")
plt.show()
```

```python
sns.boxplot(x = "Purchase Amount (USD)", y = "Frequency of Purchases" , data = good)
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency of Purchases")
plt.title("Boxplot of Purchase Amount (USD) by Frequency of Purchases")
plt.show()
```

10.Preferred Payment Method

11.Pivot table to analyze the average purchase amount by season and item purchased

```python
pd.pivot_table(good,values= "Purchase Amount (USD)", index ="Season", columns ="Item Purchased", aggfunc= "mean")
```

12.Average purchase amount by size

```python
good.groupby("Size")["Purchase Amount (USD)"].mean().sort_values( ascending = False)
```

```python
good.groupby("Size")["Purchase Amount (USD)"].mean().sort_values( ascending = False).plot(kind = "bar")
plt.xlabel("Size")
plt.ylabel("Average Purchase Amount (USD)")
plt.xticks(rotation = 0)
plt.title("Bar Chart of Average Amount by Size")
plt.show()
```

```python
ax = good.groupby("Size")["Purchase Amount (USD)"].mean().sort_values(ascending=False).plot(kind="bar")

for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="bottom"
    )

plt.show()
```
